```python
#Cleaning of the json objects as values of certain attributes like 'genres','production companies' etc
import ast
def getDataList(df, xfeature):
    xlst = []
    for x, xRows in df.iterrows():
        target_column = xRows[xfeature]
        if (target_column != 'unknown'):
            strName = ast.literal_eval( target_column )
            for i in strName:
                if(i['name'] != ''):
                    xlst.append(i['name'])
        else:
            xlst.append('unknown')
    return xlst
```

```python
#Extracting the list of genres from the 'genres' attributes. Convert from json to string.
getlist_genres = pd.DataFrame(getDataList(movies, 'genres')) 
genred=pd.DataFrame(columns=["genre","count"])

length=len(getlist_genres)

genred["genre"]=getlist_genres
genred
```

```python
# Finding the count for each unique genre from the 'genres' attribute.
pie_genre=pd.DataFrame(genred.groupby(['genre']).size())
pie_genre.rename(columns={0: 'count'},
          inplace=True, errors='raise')
pie_genre.reset_index(level=0, inplace=True)
pie_genre
```

```python
# Conversion of the count to percentage for pie chart visualisation of different genres.
pie_genre['percent']=((pie_genre['count'] / length) * 100).round(2)
pie_genre.sort_values(by=['count'])
pie_genre
```

```python
#Visualisation of different genres as a bar plot.
fig = plt.figure(figsize = (10, 5))

plt.bar(pie_genre["genre"], pie_genre["count"], color ='maroon',
        width = 0.4)
plt.xticks(rotation = 90)

# Based on the graph we draw insights that the genre 'Drama' is most watched followed by 'Comedy' and 'Thriller'.
# Thus releasing movies in these genres will lead to maximum profit to production companies.
```

```python
#Further analysis focuses on 'Cast','Directors' and 'Crew' of the movies released focusing on the 'credits' dataset.

import pandas as pd
import numpy as np
import seaborn as sns

credits2 = pd.read_csv("tmdb_5000_credits.csv")
movies2 = pd.read_csv("tmdb_5000_movies.csv")
```

```python
import json
# changing the genres column from json to string
movies2['genres'] = movies2['genres'].apply(json.loads)
for index,i in zip(movies2.index,movies2['genres']):
    list1 = []
    for j in range(len(i)):
        list1.append((i[j]['name'])) # the key 'name' contains the name of the genre
    movies2.loc[index,'genres'] = str(list1)
```

```python
# changing the keywords column from json to string
movies2['keywords'] = movies2['keywords'].apply(json.loads)
for index,i in zip(movies2.index,movies2['keywords']):
    list1 = []
    for j in range(len(i)):
        list1.append((i[j]['name']))
    movies2.loc[index,'keywords'] = str(list1)
    
```

```python
# changing the cast column from json to string
credits2['cast'] = credits2['cast'].apply(json.loads)
for index,i in zip(credits2.index,credits2['cast']):
    list1 = []
    for j in range(len(i)):
        list1.append((i[j]['name']))
    credits2.loc[index,'cast'] = str(list1)
```

```python
# changing the crew column from json to string    
credits2['crew'] = credits2['crew'].apply(json.loads)
def director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
credits2['crew'] = credits2['crew'].apply(director)
credits2.rename(columns={'crew':'director'},inplace=True)
```

```python
movies2.iloc[25]
```

```python
movies2 = movies2.merge(credits2,left_on='id',right_on='movie_id',how='left')
movies2 = movies2[['id','original_title','genres','cast','vote_average','director','keywords']]

movies2['cast'] = movies2['cast'].str.strip('[]').str.replace(' ','').str.replace("'",'').str.replace('"','')
movies2['cast'] = movies2['cast'].str.split(',')
```

```python
#Visualisation of the Actors that have the highest appearance in various movies.
#This could mean they are most liked by the public and their appearance in movies could lead to higher profits.
plt.subplots(figsize=(12,10))
list1=[]
for i in movies2['cast']:
    list1.extend(i)
ax=pd.Series(list1).value_counts()[:15].sort_values(ascending=True).plot.barh(width=0.9,color=sns.color_palette('muted',40))
for i, v in enumerate(pd.Series(list1).value_counts()[:15].sort_values(ascending=True).values): 
    ax.text(.8, i, v,fontsize=10,color='white',weight='bold')
plt.title('Actors with highest appearance')
plt.show()
```

```python
#Visualisation of the Directors that have the highest appearance in various movies.
#This could mean they are most liked by the public and their movies could lead to higher profits.
plt.subplots(figsize=(12,10))
ax = movies2[movies2['director']!=''].director.value_counts()[:10].sort_values(ascending=True).plot.barh(width=0.9,color=sns.color_palette('muted',40))
for i, v in enumerate(movies2[movies2['director']!=''].director.value_counts()[:10].sort_values(ascending=True).values): 
    ax.text(.5, i, v,fontsize=12,color='white',weight='bold')
plt.title('Directors with highest movies')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import nltk
from nltk.corpus import stopwords

#plt.subplots(figsize=(12,12))
stop_words = set(stopwords.words('english'))
stop_words.update(',',';','!','?','.','(',')','$','#','+',':','...',' ','')
```

```python
#Below is a wordcloud showing the words in the 'keywords' attributes used for describing the movies after preprocessing.

words=movies['keywords'].dropna().apply(nltk.word_tokenize)
word=[]
for i in words:
    word.extend(i)
word=pd.Series(word)
word=([i for i in word.str.lower() if i not in stop_words])
wc = WordCloud(background_color="black", max_words=2000, stopwords=STOPWORDS, max_font_size= 60,width=1000,height=1000)
wc.generate(" ".join(word))
plt.imshow(wc)
plt.axis('off')
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```
